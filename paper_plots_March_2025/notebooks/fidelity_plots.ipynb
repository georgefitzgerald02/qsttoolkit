# Model Performance (Fidelity) Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
MLE_fidelities = pd.read_csv('MLE_num_fidelities.csv')
MLE_fidelities = MLE_fidelities.drop(columns=['Unnamed: 0'])
GAN_fidelities = pd.read_csv('GAN_num_fidelities.csv')
GAN_fidelities = GAN_fidelities.drop(columns=['Unnamed: 0'])
Multitask_fidelities_1 = pd.read_csv('Multitask_num_fidelities_1.csv')
Multitask_fidelities_1 = Multitask_fidelities_1.drop(columns=['Unnamed: 0'])
Multitask_fidelities_2 = pd.read_csv('Multitask_num_fidelities_2.csv')
Multitask_fidelities_2 = Multitask_fidelities_2.drop(columns=['Unnamed: 0'])
```

```python
MLE_fidelities.head()
```

```python
GAN_fidelities.head()
```

```python
Multitask_fidelities = pd.DataFrame({
    'epochs': Multitask_fidelities_1['epochs'],
    'num_fidelities_0': Multitask_fidelities_1['num_fidelities_0'],
    'num_times_0': Multitask_fidelities_1['num_times_0'],
    'num_fidelities_1': Multitask_fidelities_1['num_fidelities_1'],
    'num_times_1': Multitask_fidelities_1['num_times_1'],
    'num_fidelities_2': Multitask_fidelities_1['num_fidelities_2'],
    'num_times_2': Multitask_fidelities_1['num_times_2'],
    'num_fidelities_3': Multitask_fidelities_2['num_fidelities_0'],
    'num_times_3': Multitask_fidelities_2['num_times_0'],
    'num_fidelities_4': Multitask_fidelities_2['num_fidelities_1'],
    'num_times_4': Multitask_fidelities_2['num_times_1']
})
Multitask_fidelities.head()
```

```python
MLE_fidelities['num_average'] = MLE_fidelities.filter(regex='num_fidelities').mean(axis=1)
MLE_fidelities['num_stderr'] = MLE_fidelities.filter(regex='num_fidelities').std(axis=1)#.apply(lambda x: x/np.sqrt(5))
MLE_fidelities.head(20)
```

```python
GAN_fidelities['num_average'] = GAN_fidelities.filter(regex='num_fidelities').mean(axis=1)
GAN_fidelities['num_stderr'] = GAN_fidelities.filter(regex='num_fidelities').std(axis=1)#.apply(lambda x: x/np.sqrt(5))
GAN_fidelities.loc[0, 'epochs'] = 0.1
GAN_fidelities.head(20)
```

```python
Multitask_fidelities = Multitask_fidelities.drop(columns=Multitask_fidelities.filter(regex='cat').columns)

# Multitask_zeroth = pd.read_csv('Multitask_zeroth_fidelities.csv')
# Multitask_zeroth = Multitask_zeroth.drop(columns=['Unnamed: 0'])

# Multitask_fidelities = pd.concat([Multitask_zeroth.transpose().rename(columns=lambda x: 'num_fidelities_' + str(x)), Multitask_fidelities], ignore_index=True)
# Multitask_fidelities.loc[0, 'epochs'] = 0.1

Multitask_fidelities['num_average'] = Multitask_fidelities.filter(regex='num_fidelities').mean(axis=1)
# Multitask_fidelities['cat_average'] = Multitask_fidelities.filter(regex='cat').mean(axis=1)
Multitask_fidelities['num_stderr'] = Multitask_fidelities.filter(regex='num_fidelities').std(axis=1)#.apply(lambda x: x/np.sqrt(5))
# Multitask_fidelities['cat_stderr'] = Multitask_fidelities.filter(regex='cat').std(axis=1).apply(lambda x: x/np.sqrt(5))

Multitask_fidelities.head(20)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(MLE_fidelities['epochs'], MLE_fidelities['num_average'], label='MLE', color='#68246D')
axs[0].fill_between(MLE_fidelities['epochs'], MLE_fidelities['num_average'] - MLE_fidelities['num_stderr'], MLE_fidelities['num_average'] + MLE_fidelities['num_stderr'], alpha=0.2, color='#68246D')

axs[0].plot(GAN_fidelities['epochs'], GAN_fidelities['num_average'], label='GAN', color='green')
axs[0].fill_between(GAN_fidelities['epochs'], GAN_fidelities['num_average'] - GAN_fidelities['num_stderr'], GAN_fidelities['num_average'] + GAN_fidelities['num_stderr'], alpha=0.2, color='green')

axs[0].plot(Multitask_fidelities['epochs'], Multitask_fidelities['num_average'], label='Multitask', color='orange')
axs[0].fill_between(Multitask_fidelities['epochs'], Multitask_fidelities['num_average'] - Multitask_fidelities['num_stderr'], Multitask_fidelities['num_average'] + Multitask_fidelities['num_stderr'], alpha=0.2, color='orange')

# ax.set_title('Num States', fontsize=20)
axs[0].set_xlabel('Epoch', fontsize=12)
axs[0].set_ylabel('Fidelity', fontsize=12)
axs[0].set_xscale('log')
axs[0].set_xlim([1, 1000])
axs[0].set_ylim([0, 1])
axs[0].legend()


for run in range(5):
    axs[1].plot(MLE_fidelities['num_times_' + str(run)], MLE_fidelities['num_fidelities_' + str(run)], color='#68246D', linewidth=.5)
    axs[1].plot(GAN_fidelities['num_times_' + str(run)], GAN_fidelities['num_fidelities_' + str(run)], color='green', linewidth=.5)
    axs[1].plot(Multitask_fidelities['num_times_' + str(run)], Multitask_fidelities['num_fidelities_' + str(run)], color='orange', linewidth=.5)

# Add legend entries manually
axs[1].plot([], [], color='#68246D', label='MLE')
axs[1].plot([], [], color='green', label='GAN')
axs[1].plot([], [], color='orange', label='Multitask')

axs[1].set_xlabel('Time (s)', fontsize=12)
axs[1].set_ylabel('Fidelity', fontsize=12)
axs[1].set_xscale('log')
axs[1].set_xlim([1, 1000])
axs[1].set_ylim([0, 1])
axs[1].legend()

plt.savefig('comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 4))
ax.plot(MLE_fidelities['epochs'], MLE_fidelities['num_average'], label='MLE', color='#68246D')
ax.fill_between(MLE_fidelities['epochs'], MLE_fidelities['num_average'] - MLE_fidelities['num_stderr'], MLE_fidelities['num_average'] + MLE_fidelities['num_stderr'], alpha=0.2, color='#68246D')

ax.plot(GAN_fidelities['epochs'], GAN_fidelities['num_average'], label='GAN', color='green')
ax.fill_between(GAN_fidelities['epochs'], GAN_fidelities['num_average'] - GAN_fidelities['num_stderr'], GAN_fidelities['num_average'] + GAN_fidelities['num_stderr'], alpha=0.2, color='green')

ax.plot(Multitask_fidelities['epochs'], Multitask_fidelities['num_average'], label='Multitask', color='orange')
ax.fill_between(Multitask_fidelities['epochs'], Multitask_fidelities['num_average'] - Multitask_fidelities['num_stderr'], Multitask_fidelities['num_average'] + Multitask_fidelities['num_stderr'], alpha=0.2, color='orange')

# ax.set_title('Num States', fontsize=20)
ax.set_xlabel('Epoch', fontsize=12)
ax.set_ylabel('Fidelity', fontsize=12)
ax.set_xscale('log')
ax.set_xlim([1, 1000])
ax.set_ylim([0, 1])
ax.legend()

plt.savefig('comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
